### Dados do UCI

```python
from sklearn.datasets import load_wine

# Carregando o dataset 'Wine'
wine_data = load_wine()
df_wine = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)

# Criando um dicionário para mapear os nomes das colunas para português
colunas_pt = {
    'alcohol': 'alcool',
    'malic_acid': 'acido_malico',
    'ash': 'cinza',
    'alcalinity_of_ash': 'alcalinidade_da_cinza',
    'magnesium': 'magnesio',
    'total_phenols': 'fenois_totais',
    'flavanoids': 'flavonoides',
    'nonflavanoid_phenols': 'fenois_nao_flavonoides',
    'proanthocyanins': 'proantocianidinas',
    'color_intensity': 'intensidade_da_cor',
    'hue': 'matiz',
    'od280/od315_of_diluted_wines': 'od280/od315_de_vinhos_diluídos',
    'proline': 'prolina'
}

# Renomeando as colunas para português
df_wine.rename(columns=colunas_pt, inplace=True)

# Exibindo as primeiras linhas do dataset
print("Primeiras linhas do dataset Wine:")
df_wine.head()
```
